###4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a dataset with categories and numerical values
data = {
    'Category': np.random.choice(['A', 'B', 'C'], size=100),
    'Value': np.random.randint(10, 50, size=100)
}

# Create a DataFrame from the dataset
df = pd.DataFrame(data)

# Group the data by category and calculate the mean value
grouped_data = df.groupby('Category')['Value'].mean()

# Create a bar chart to visualize the distribution
plt.bar(grouped_data.index, grouped_data.values)
plt.xlabel('Category')
plt.ylabel('Mean Value')
plt.title('Distribution of Values Across Categories')
plt.show()
```

###5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a synthetic dataset with correlated features
np.random.seed(42)
x1 = np.random.randn(100)
x2 = x1 + 0.5 * np.random.randn(100)
x3 = x1 + 0.3 * np.random.randn(100)
x4 = x2 + 0.2 * np.random.randn(100)

# Create a DataFrame from the data
df = pd.DataFrame({'x1': x1, 'x2': x2, 'x3': x3, 'x4': x4})

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap to visualize the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='viridis')

# Set the title
plt.title('Correlation Matrix')

# Show the plot
plt.show()
```

###1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a three- dimensional space

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate the dataset
np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

# Create a 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the data points
ax.scatter(df['X'], df['Y'], df['Z'])

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Scatter Plot')

# Show the plot
plt.show()
```

###2. Using the Student Grades, create a violin plot to display the distribution of scores across different grade categories.


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Student Grades Data
np.random.seed(15)
data_grades = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df_grades = pd.DataFrame(data_grades)

# Sales Data
np.random.seed(20)
data_sales = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df_sales = pd.DataFrame(data_sales)

# Violin plot for Student Grades
sns.violinplot(x='Grade', y='Score', data=df_grades)
plt.title('Distribution of Scores by Grade')
plt.xlabel('Grade')
plt.ylabel('Score')
plt.show()

# Heatmap for Sales Data
sns.heatmap(df_sales.pivot_table(index='Month', columns='Day', values='Sales'))
plt.title('Sales Variation by Month and Day')
plt.xlabel('Day')
plt.ylabel('Month')
plt.show()
```

###3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Create a pivot table to organize the data by Month and Day
pivot_table = df.pivot_table(index='Month', columns='Day', values='Sales')

# Generate a heatmap
sns.heatmap(pivot_table, cmap='viridis')

# Set labels and title
plt.xlabel('Day')
plt.ylabel('Month')
plt.title('Sales Variation by Month and Day')

# Show the plot
plt.show()
```

###4. Using the given x and y data, generate a 3D surface plot to visualize the function z = sin(sqrt(x^2 + y^2))

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate the data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, z, cmap='viridis', alpha=0.8)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot of sin(sqrt(x^2 + y^2))')

# Show the plot
plt.show()
```

###5. Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create a bubble chart
plt.scatter(df['GDP'], df['Population'], s=df['Population'] * 10, alpha=0.7)

# Add labels and a title
plt.xlabel('GDP')
plt.ylabel('Population')
plt.title('Bubble Chart of Population and GDP')

# Add annotations
for i, row in df.iterrows():
    plt.annotate(row['Country'], (row['GDP'], row['Population']))

# Show the plot
plt.show()
```

###1. Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import Grid, Label, Title
import numpy as np

# Ensure Bokeh outputs to the notebook
output_notebook()

# Create a figure
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')

# Generate x and y values
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot the line
p.line(x, y, color='blue', line_width=2)

# Add grid lines
p.add_layout(Grid(grid_line_color="gray", grid_line_alpha=0.2))

# Show the plot
show(p)
```

###2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
import numpy as np

# Generate random data
x = np.random.randn(100)
y = np.random.randn(100)
sizes = np.random.randint(10, 50, 100)
colors = np.random.rand(100, 3)  # Random RGB colors

# Create a ColumnDataSource
source = ColumnDataSource(data={'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a figure
p = figure(title="Scatter Plot with Sizes and Colors")

# Plot the scatter points
p.scatter('x', 'y', source=source, size='sizes', color='colors', alpha=0.7)

# Show the plot
show(p)
```

###3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset

    fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
    counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral11

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

output_file("fruit_counts.html")

p = figure(x_range=fruits, height=400, title="Fruit Counts",
           toolbar_location=None, tools="")

p.vbar(x=fruits, top=counts, width=0.9, line_color='white',
       fill_color=factor_cmap('x', palette=Spectral11, factors=fruits))

p.xgrid.grid_line_color = None
p.y_range.start = 0
p.yaxis.axis_label = "Counts"
p.xaxis.axis_label = "Fruits"
p.xaxis.major_label_orientation = "vertical"

show(p)
```

###4. Create a Bokeh histogram to visualize the distribution of the given data.
    data_hist = np.random.randn(1000)
    hist, edges = np.histogram(data_hist, bins=30)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_file

data_hist = np.random.randn(1000)

hist, edges = np.histogram(data_hist, bins=30)

output_file("histogram.html")

p = figure(title="Histogram of Normally Distributed Data",
           x_axis_label='Value',
           y_axis_label='Frequency',
           tools="")

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='navy', line_color='white')

show(p)
```

###5. Create a Bokeh heatmap using the provided dataset.
    data_heatmap = np.random.rand(10, 10)
    x = np.linspace(0, 1, 10)
    y = np.linspace(0, 1, 10)
    xx, yy = np.meshgrid(x, y)


```python
from bokeh.plotting import figure, show
import numpy as np
from bokeh.palettes import Viridis256 # Import the Viridis256 palette

# Generate random data
data_heatmap = np.random.rand(10, 10)

# Create x and y coordinates
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create a figure
p = figure(title="Heatmap", x_axis_label="X", y_axis_label="Y")

# Create a heatmap
# The image argument must be a 1D array or a list of 1D arrays
# Use list(Viridis256) to convert the palette to a list of colors
p.image(image=[data_heatmap], x=xx[0, 0], y=yy[0, 0], dw=1, dh=1, palette=list(Viridis256))

# Show the plot
show(p)
```
