###3. Explain the following terms

###The difference in np.array, np.asarray and np.asanyarray

###The difference between Deep copy and shallow copy

**np.array():**

This function creates a new array object from any input data (like a list, tuple, etc.). It always makes a copy of the input data by default unless specified otherwise. Even if the input is already an ndarray, np.array() can create a new array based on the copy parameter.

```python
import numpy as np

a = [1, 2, 3]
b = np.array(a)

b
```

**np.asarray():**

This function is similar to np.array(), but it tries to avoid copying the data if the input is already a NumPy array. It will convert the input to an array if it’s not already, but it will not create a new copy if the input is already a NumPy array.

```python
a = np.array([1, 2, 3])
b = np.asarray(a)

b
```

**np.asanyarray():**

This function is similar to np.asarray(), but with an important difference: it will not force the input to become a base ndarray if it’s a subclass of ndarray. For example, if the input is a matrix or masked array, np.asanyarray() will keep it as that specific subclass, whereas np.asarray() would convert it to a base ndarray.

```python
a = np.matrix([[1, 2], [3, 4]])
b = np.asanyarray(a)

b
```

**Shallow Copy:**

A shallow copy creates a new object, but does not copy nested objects. Instead, it copies the references to the original nested objects. This means that changes to nested objects in the copied object will affect the original object as well, because both objects are referencing the same nested elements.
In NumPy, a shallow copy can be made using functions like np.view() or np.asarray() if the input is already an array.

```python
import copy
a = [1, 2, [3, 4]]
b = copy.copy(a)  # shallow copy
b[2][0] = 100
print(a)
```

Deep Copy:

A deep copy creates a completely independent new object, including all nested objects. Changes made to the copied object do not affect the original one. Every element, including nested structures, is recursively copied.

In NumPy, a deep copy can be made using the copy() method on an array.

```python
import copy
a = [1, 2, [3, 4]]
b = copy.deepcopy(a)  # deep copy
b[2][0] = 100
print(a)
```

###14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:

a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a DataFrame with 'Date', 'Temperature', and 'Humidity' columns
start_date = pd.to_datetime('2023-01-01')
end_date = pd.to_datetime('2023-12-31')
dates = pd.date_range(start_date, end_date, freq='D')

temperature_data = np.random.randint(50, 100, size=len(dates))
humidity_data = np.random.randint(30, 90, size=len(dates))

df = pd.DataFrame({'Date': dates, 'Temperature': temperature_data, 'Humidity': humidity_data})

# Set 'Date' as the index
df.set_index('Date', inplace=True)

# Create the plot
fig, ax1 = plt.subplots()

# Plot Temperature on the left y-axis
ax1.plot(df['Temperature'], color='blue', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (Celsius)', color='blue')

# Create a second axis for Humidity on the right y-axis
ax2 = ax1.twinx()
ax2.plot(df['Humidity'], color='red', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='red')

# Set the title
plt.title('Temperature and Humidity Over Time')

# Show the legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

###15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:

a) Plot a histogram of the data with 30 bins.

b) Overlay a line plot representing the normal distribution's probability density function (PDF).

c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

d) Set the title of the plot as 'Histogram with PDF Overlay'.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create a NumPy array with 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot a histogram with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, label='Histogram')

# Get the mean and standard deviation of the data
mean, std = data.mean(), data.std()

# Generate x values for the PDF
x = np.linspace(min(data), max(data), 100)

# Calculate the PDF values
pdf = norm.pdf(x, mean, std)

# Plot the PDF
plt.plot(x, pdf, color='red', label='PDF')

# Label the axes
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Set the title
plt.title('Histogram with PDF Overlay')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```

###17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Create two random arrays
x = np.random.randn(500)
y = np.random.randn(500)

# Determine the quadrant for each point
quadrants = np.where(x >= 0, np.where(y >= 0, 1, 4), np.where(y >= 0, 2, 3))

# Create a scatter plot with colors based on quadrants
sns.scatterplot(x=x, y=y, hue=quadrants, palette='viridis')

# Add a legend
plt.legend(title='Quadrant')

# Label the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set the title
plt.title('Quadrant-wise Scatter Plot')

# Show the plot
plt.show()
```

###18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import Grid
import numpy as np

# Ensure Bokeh outputs to the notebook
output_notebook()

# Create a figure
plt = figure(title="Sine Wave Function", x_axis_label='x', y_axis_label='y')

# Create a sine wave function
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

# Plot the line
plt.line(x, y, color='blue', line_width=2)

# Add grid lines
plt.add_layout(Grid(grid_line_color="gray", grid_line_alpha=0.2))

# Show the plot
show(plt)
```

###19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.

```python
from bokeh.plotting import figure, show
from bokeh.models import Label, Title, HoverTool
import numpy as np
from bokeh.palettes import viridis # Import the viridis palette from bokeh.palettes

# Generate random categorical data
categories = np.random.choice(['A', 'B', 'C', 'D', 'E'], size=100)

# Convert categories to integers
unique_categories, counts = np.unique(categories, return_counts=True)

# Create a figure
p = figure(title="Random Categorical Bar Chart", x_axis_label='Category', y_axis_label='Count')

# Create a bar chart using unique categories as x-axis labels
p.vbar(x=unique_categories, width=0.8, top=counts, color=viridis(len(counts))) # Call the viridis function to get the palette

# Add hover tooltips
hover = HoverTool(tooltips=[('Category', '@x'), ('Count', '@top')], mode='vline')
p.add_tools(hover)

# Show the plot
show(p)
```
